# 18강- 적률생성함수\_2 (MGFs Continued) 

### 확률분포의 적률생성함수 및 적률 예제  
  
**지수분포**
X ~ Expo(1)일 때, MGF와 Moment를 구하라.  
  
MGF    
![](https://latex.codecogs.com/gif.latex?%5C%5CM%28t%29%3DE%28e%5E%7BtX%7D%29%5C%5C%20%3D%20%5Cdisplaystyle%20%5Cint%5E%5Cinfty%20_0%20e%5E%7BtX%7De%5E%7B-x%7Ddx%5C%20%28by%5C%20LOTUS%29%5C%5C%20%3D%20%5Cdisplaystyle%20%5Cint%5E%5Cinfty%20_0%20e%5E%7B-x%281-t%29%7Ddx%5C%5C%20%3D%20%5Cfrac%20%7B1%7D%7B1-t%7D%20%2C%5C%20t%20%3C%201%5C%5C)  
  
E(e^(tX))는 LOTUS를 이용해서 위의 식이 나오며, 적분을 풀기위해 식을 다른 파라미터를 가진 지수 확률밀도함수로 보고, 정규화상수를 넣어서 풀면 1/t-t가 되며,  
t ≥ 1일 경우 적분값이 ∞가 되어버려, 문제가 생기기 때문에 t < 1이라는 조건을 준다. 
이의 경우 t < 1에서 유한한 특정 구간을 취할 수 있기 때문에 완벽한 MGF이다. 
  
Moment
![](https://latex.codecogs.com/gif.latex?%5C%5C%20M%27%280%29%20%3D%20E%28X%29%20%5C%5C%20M%27%27%280%29%20%3DE%28X%5E2%29%20%5C%5C%20%5Ccdots%20%5C%5C%20When%5C%20%7Ct%7C%5C%20%3C1%2C%20%5C%5C%5Cfrac%7B1%7D%7B1-t%7D%20%3D%20%5Csum%20%5E%5Cinfty%20_%7Bn%3D1%7Dt%5En%28%5Cbecause%5C%20geometric%5C%20series%29%20%3D%20%5Csum%20%5E%5Cinfty%20_%7Bn%3D0%7D%20n%21%20%5Cfrac%7Bt%5En%7D%7Bn%21%7D%20%5C%5C%5C%5C%20%5Crightarrow%20E%28X%5En%29%20%3D%20n%21%5C%20%28%5Cbecause%20coefficient%5C%20of%5C%20%5Cfrac%7Bt%5En%7D%7Bn%21%7D%5C%20is%5C%20moment%29%20%5C%5C%20When%5C%20Y%20%5Csim%20Expo%28%5Clambda%29%2C%5C%20X%20%3D%20%5Clambda%20Y%2C%5C%5C%20Y%5En%20%3D%20%5Cfrac%7BX%5En%7D%7B%5Clambda%20%5En%7D%5C%5C%20%5Crightarrow%5C%20E%28Y%5En%29%20%3D%5Cfrac%20%7Bn%21%7D%7B%5Clambda%5En%7D)  
  
이 전의 강의에서 말했듯이 적률은 t^n / n!의 계수이며,  
여기서는 미적분을 사용하지 않고, 기하급수만을 이용해서 적률을 구할 수 있었다.   
  
**정규분포**  
    
Z ~ N(0,1) 일 때, MGF와 MOMENT를 구하라.  
  
MGF  
  
일단 E(Z) = 0, E(Z^2) = 1 임을 알 수 있으며,  
n이 홀수(odd)일 때, E(Z^n) = 0 이다. (∵ 대칭성)  
  
그러나 짝수의 경우는 구하기가 힘들다.  
  
![](https://latex.codecogs.com/gif.latex?%5C%5C%20M%28t%29%20%3D%20e%5E%7Bt%5E2/2%7D%3D%20%5Cdisplaystyle%20%5Csum%20%5E%5Cinfty%20_%7Bn%3D0%7D%20%5Cfrac%7B%28t%5E2/2%29%5En%7D%7Bn%21%7D%5C%20%28%5Cbecause%20taylor%5C%20series%29%3D%20%5Csum%5E%5Cinfty%20_%7Bn%3D0%7D%20%5Cfrac%7Bt%5E%7B2n%7D%7D%7B2%5En%20n%21%7D%20%3D%20%5Csum%20%5E%5Cinfty%20_%7Bn%3D0%7D%20%5Cfrac%7B%282n%29%21t%5E%7B2n%7D%7D%7B2%5En%20n%21%20%282n%29%21%7D%20%5C%5C%5CRightarrow%20E%28Z%5E%7B2n%7D%29%20%3D%20%5CLarge%20%5Cfrac%7B%282n%29%21%7D%7B2%5En%20n%21%7D%20%5C%5C%20E%28Z%5E2%29%20%3D%201%20%5C%5C%20E%28Z%5E4%29%20%3D%201%20%5Ccdot%203%20%3D%203%20%5C%5C%20E%28Z%5E6%29%20%3D%201%5Ccdot%203%20%5Ccdot%205%20%3D%2015)  
  
e^{t^2 / 2}는 e^x의 테일러 급수가 어디에서나 수렴하므로 테일러 급수를 사용한다.  
그리고 이 함수는 우함수이므로, t의 짝수 지수만 나올 것이며, t^2n으로 계산을 한 것이다.  위의 MGF는 2n명 사람이 n짝을 이루는 경우의 수를 구하는 식과 같다.  
  
**포아송 분포**  
  
포아송분포의 평균, 분산은 모두 λ이다.    
  
X ~ Pois(λ)일 때, MGF를 구하라.  
  
![](https://latex.codecogs.com/gif.latex?%5C%5CE%28e%5E%7BtX%7D%29%20%5C%5C%3D%20%5Cdisplaystyle%20%5Csum%20%5E%5Cinfty%20_%7Bk%3D0%7D%20%5Cfrac%7Be%5E%7Btk%7D%20e%5E%7B-%20%5Clambda%7D%20%5Clambda%20%5Ek%7D%7Bk%21%7D%20%28e%5E%7Btk%7D%5C%20by%5C%20LOTUS%29%20%5C%5C%5C%5C%3D%20e%5E%7B-%20%5Clambda%7De%5E%7B%5Clambda%20e%5Et%7D%28%5Cbecause%20%5Cdisplaystyle%20%5Csum%5E%5Cinfty%20_%7Bk%3D1%7D%20%5Cfrac%7B%28e%5Et%20%5Clambda%29%5Ek%7D%7Bk%21%7D%20%3D%20e%5E%7B%5Clambda%20e%5Et%7D%5C%20%3A%5C%20taylor%5C%20series%29%20%5C%5C%3D%20e%5E%7B%5Clambda%28e%5Et-1%29%7D)  
  
e^tk는 LOTUS에 의해서 나왔으며, 포아송분포의 확률질량함수를 곱해준다.  
e^{-λ}는 상수라 앞으로 나오며, 남은 것을 합치면 (λe^{t})^{k} / k!의 합이며,  
이는 e^x의 테일러 급수가 된다.(x에 λe^t를 대입한 e^x의 테일러 급수)  
여기서 나온 MGF는 모든 t에 대해 유효하다.(모든 t에 대해서 급수가 수렴하기 때문)  

X, Y가 독립이며, 각각 포아송분포를 따를 때, X+Y의 분포는?  
  
적분을 사용하면 굉장히 어려울 것이지만  
여기서는 X의 MGF와 Y의 MGF를 곱해주기만 하면 답이 나온다.  
  
![](https://latex.codecogs.com/gif.latex?%5C%5C%20When%5C%20X%20%5Csim%20Pois%28%5Clambda%29%2C%5C%20Y%20%5Csim%20Pois%28%5Cmu%29%2C%20X%5C%20and%5C%20Y%5C%20are%5C%20independent.%20%5C%5C%20Find%5C%20distribution%5C%20of%5C%20X&plus;Y.%5C%20%28%3DConvolution%29%20%5C%5C%5C%5C%20Multiply%5C%20MGFs%20%3A%5C%20e%5E%7B%5Clambda%28e%5Et-1%29%7De%5E%7B%5Cmu%28e%5Et-1%29%7D%20%3D%20e%5E%7B%28%5Clambda%20&plus;%20%5Cmu%29%28e%5Et-1%29%7D%5CRightarrow%20X&plus;Y%20%5Csim%20Pois%28%5Clambda%20&plus;%20%5Cmu%29%u21D2X&plus;Y%u223CPois%28%u03BB&plus;%u03BC%29)  
  
포아송분포의 경우 선형성에 따라서 독립적인 두 분포를 더하면 그대로 포아송분포이다.  
(다른 분포에서는 볼 수 없음.)  
  
만약 X와 Y가 종속관계일 경우에는(X=Y일 경우를 한번 보자)  
평균과 분산을 통해 간단하게 알 수 있다.  
X+Y = 2X이며, E(X) = 2λ가 되고, Var(2X) = 4λ가 되어, 포아송분포가 되지 않을 것이다.  
  
### 결합분포(joint distribution)  
  
한 개 이상의 확률변수를 가지는 분포를 어떻게 다룰 것인지를 보여준다.  
  
확률변수 X, Y가 있다.  
  
joint CDF(결합 누적분포함수):    
F(x, y) = P(X ≤ x, Y ≤ y)  
joint PMF(결합 확률질량함수)(이산적일 때(discrete)):  
P(X = x, Y = y)  
joint PDF(결합 확률밀도함수)(이차원 공간에서 연속적일 때(continuous)):  
일단 B를 어떤 평면의 면적을 뜻하는 집합이라고 할 때,  
![](https://latex.codecogs.com/gif.latex?%5C%5C%20f%28x%2Cy%29%5C%20such%5C%20that%5C%20P%28%28x%2Cy%29%20%5Cin%20B%29%20%3D%20%5Cdisplaystyle%20%5Ciint%20_B%20f%28x%2Cy%29dxdy)  

**주변 분포(marginal distribution)**  
확률변수 X의 marginal CDF(주변 누적분포함수):  f(x) = P(X ≤ x)     
  
독립일 때 결합 누적분포함수는 주변 누적분포함수의 곱이다.  
**독립**    
![](https://latex.codecogs.com/gif.latex?X%2CY%5C%20are%5C%20independent%20%5CLeftrightarrow%20F%28x%2Cy%29%20%3D%20F_X%28x%29F_Y%28y%29%20%5C%5C%20%5CRightarrow%20discrete%29%20%5C%20P%28X%3Dx%2C%20Y%3Dy%29%20%3D%20P%28X%3Dx%29P%28Y%3Dy%29%20%5C%5Ccontinuous%29%5C%20f%28x%2Cy%29%20%3D%20f_X%28x%29f_Y%28y%29%2C%5C%20for%5C%20all%5C%20%28x%2Cy%29%20%5Cin%20%5CLarge%20R)

ex1) 이산적인 경우   
X, Y는 베르누이 분포를 따른다.  
 
| | Y = 0 | Y = 1 | 
|:--------|:--------:|--------:|
| **X = 0** | 2/6 | 1/6 |   
| **X = 1** | 2/6 | 1/6 |   
이렇게 임의로 확률을 줬을 때 X, Y는 서로 독립적인 관계인가?    
  
이를 위해서 주변 누적분포함수를 구해서 곱한다음 결합 누적분포함수와 비교해보자.  
  
**결합분포에서 주변분포를 구하자.**
  
![](https://latex.codecogs.com/gif.latex?%5C%5C%20discrete%29%5C%20P%28X%3Dx%29%20%3D%20%5Cdisplaystyle%20%5Csum%20_y%20P%28X%20%3D%20x%2CY%20%3D%20y%29%20%5C%5C%20continuous%29%5C%20f_X%28y%29%20%3D%20%5Cdisplaystyle%20%5Cint%20%5E%5Cinfty%20_%7B-%5Cinfty%7D%20f_%7BX%2CY%7D%28x%2Cy%29dy)  
  
반대로 주변분포에서 결합분포로 구할 수는 없다!! → 주변분포만 알아서는 X,Y의 관계를 알 수 없기 때문!!  
    
![](https://cphinf.pstatic.net/mooc/20180902_260/1535881716124eeHco_PNG/18-1.png)

![](https://cphinf.pstatic.net/mooc/20180902_146/1535881861915BPWyi_PNG/18-2.png)  
위의 경우 주변 누적분포함수의 곱 ≠ 결합 누적분포함수 이므로 종속이다!
