##### law of total probability(전체 확률의 법칙)   
A_1, ... , A_n 까지를 S의 분할이라고 해보자.(=각각이 서로소이고 다 합하면 S가 되는 것)   
그렇다면  
P(B) = P(B∩A_1) + ... + P(B∩A_n)  
     = P(B|A_1)P(A_1) + ... + P(B|A_n)P(A_n)  

조건부확률은 그 자체로도 중요하기도 하지만,
위의 P(B)와 같은 비조건부확률을 작은 문제로 쪼갤 경우에는 P(B|A_1)과 같은 조건부 확률이 필요하다.  
  
ex) 일반적인 52장의 카드 중에 임의로 2장의 카드를 뽑는다고 가정하자.  
하나의 카드가 A라고 주어 졌을 경우, 2장의 카드가 모두 A일 확률을 구해보자.  
  
P(2장의 에이스 | 에이스를 1장 가졌을 때)  
= P(2장 에이스, ~~1장 에이스~~) **/** P(1장 에이스)  
**->2장이 에이스일 경우 1장 에이스인 경우를 이미 포함하기 때문에 불필요한 '1장 에이스'인 조건을 지운다.**  
= (4 C 2 / 52 C 2) **/** (1 - 48 C 2 / 52 C 2)  
**-> '1장 에이스'일 경우는 2장의 에이스를 가진 경우와 1장은 에이스 나머지 1장은 에이스가 아닌 것을 가질 경우로 나누어 질 수 있다.(여사건을 활용해서 1장도 에이스를 가지지 못할 경우를 빼준다.)**  
= 1 / 33  
  
P(2장의 에이스 | 스페이드 A를 1장 가졌을 때) (= 1장의 스페이드를 가졌을 때 나머지 1장의 카드가 에이스)  
 = 3 / 51 (52장 중에 이미 1장을 골라 스페이드 A가 나왔고, 카드가 나오는 확률이 대칭적이기 때문에, 스페이드 A를 제외한 51장 중에 클로버, 다이아, 하트 A 3장이 나올 확률)  
 = 1 / 17  

스페이드라고 모양을 지정해 주었을 뿐인데 거의 2배의 확률 차이가 난다.  
  
**그 이유는 말하는 카드가 최소한 하나인지 특정한 하나(스페이드 A)인지에서 오는 차이이다.**  
  
환자가 들어와서 특정 질병 여부를 검사한다고 가정해보자.  
이 질병은 인구의 1%가 앓고 있으며, 해당 환자가 질병검사에 대해서 양성반응이 나왔다고 하자.  
해당 질병 검사는 95%의 정확성을 가지고 있다.  
  
D : 환자가 질병을 가지는 사건  
T : 해당 환자가 질병검사에 양성 판정을 받는 사건  
  
P(T | D) = 0.95 = P(T^c | D^c)  
 = 환자가 질병을 가졌다면 해당 환자가 정확하게 양성 판정 받을 확률  
 = 환자가 질병을 가지고 있지 않다면 해당 환자가 정확하게 음성 판정 받을 확률  

그러나 환자가 원하는 것은 P(T | D)가 아니라 P(D | T) 일 것이다.
즉, 질병검사에 양성 판정을 받았을 경우 질병에 걸렸을 확률을 더 원할 것이다.  

P(D | T)  
= P(T | D)ㆍP(D) / P(T)__(Bayes' Theorem)__  
= P(T | D)ㆍP(D) / P(T | D)ㆍP(D) + P(T | D^c)ㆍP(D^c)__(전체 확률의 법칙)__ 
질병에 걸렸을 경우와 걸리지 않았을 경우 2가지로 분할해서 나타냄.  
≒ 0.16  
  
직관과는 다른 결과가 나온다.  
95%의 정확성이라는 수치를 보고 높게 생각할 수 있지만, 그것은 인구의 1%만 질병에 걸린다는 수치를 간과한 것.    

Biohazards(생물학적 위험요소)

#### (1) P(A|B)와 P(B|A)를 혼동하는 것.(ex) 검사의 오류)  
ex) Sally Clark case  
SIDS(영아돌연사증후군)으로 두 아이를 잃은 Sally Clark.  
그리고 그녀는 살인죄로 유죄를 선고받는다.  

전문가는 P(아이가 죽을 확률 | 결백) = 1 / 8500이라고 주장.
그리고 두 아이가 죽었고 이를 독립이라고 가정 하여  
 P(아이가 죽을 확률 | 결백) ㆍ P(아이가 뚜렷한 이유없이 죽을 확률 | 결백) = 1 / 73 ㆍ 10^6  
이는 P(증거|결백함)이다.   
  
하지만 우리가 구해야 하는 것은 P(결백 | 증거)이지 그 반대가 아니다!  

#### (2) 사전(prior) 확률인 P(A)와 사후(posterior) 확률인 P(A|B) 사이를 혼동하는 것
ex) 어떠한 문제에서 A가 일어난 후라고 주어졌을 때  
P(A) = 1이라고 생각할 수 있지만 이는 틀렸다.  
P(A|A) = 1인 것이다.  
#### (3) 독립과 조건부 독립을 혼동하는 것  
조건부 독립(conditional independence)  
Definition : 사건 A,B가 조건 C에 하에서 조건부 독립일 때  
             P(A ∩ B | C) = P(A | C)ㆍP(B | C)  
  
**조건 C하에서의 조건부 독립이 (비조건부)독립을 의미하는가? NO**  
  
조건부 독립이라면 독립 일 수도 아닐 수 도 있다.  
  
ex) 알 수 없는 상대와 체스를 둔다고 가정.  
체스를 몇 판 둔다고 하자.
만약 다섯 판을 이겼다고 생각을 해보면, 다시 붙을 때 상대방보다 더 잘 둘 수 있다고 생각할 것이다.  
즉, 이전 경기들이 상대의 실력을 가늠할 수 있는 정보를 주게 된 것이다.  
이것이 다음 결과를 예측하는데 관계가 있다.  
그래서 경기들은 상대방의 실력이 주어졌을 때 조건부 독립일 것이다.  
그래서 결과는 조건부 독립이지만 독립은 아니다.  
  
**아무런 조건없이 A,B가 독립일 때 조건 C가 주어진다면 조건부 독립이라고 할 수 있는가? NO**  
  
ex) A : 화재 경보가 울리는 사건
화재 경보는 2가지 방법에 의해서 울리게 된다.  
(1) F : 불이 나는 사건  
(2) P : 팝콘을 튀기는 사건  
그리고 F와 P는 독립이라고 가정.  
그러나 P(F | A,P^c) = 1   
이는 F와 P는 독립이지만 화재 경보가 울렸다는 조건 하에서의 조건부 독립은 아닌 것이다.  
화재 경보가 울리고 이를 설명하려고 하면 이 둘은 종속적인 관계가 된다.  
팝콘을 안 튀겼다면, 불이 났다는 것.
