#### 이산(discrete)  
확률 변수 X  
PMF(확률질량함수) P(X = x)  
CDF(누적분포함수) F\_x(x) = P(X ≤ x) (이산분포에서의 누적분포함수는 계단식 → 다루기 힘듬)  
E(X) = ∑xP(X=x)  
Var(X) = E(X^2) - (E(X))^2  
LOTUS  
![](https://latex.codecogs.com/gif.latex?E%28g%28X%29%29%20%3D%20%5Csum_%7Bx%7Dg%28x%29P%28X%3Dx%29)

#### 연속(continuous)  
확률 변수 X  
PDF(확률밀도함수) f_x(x) [P(X = x) = 0(연속분포에서는 0과 1사이에 무한한 실수가 존재 하므로, 특정한 값에 대한 확률은 0이 된다.)]  
CDF(누적분포함수) F_x(x) = P(X ≤ x) (누적분포함수는 일반적이라 모든 확률변수가 누적분포함수를 가지며, 연속분포의 경우 이산분포와 달리 누적분포함수는 이어져 있다.)  
**확률밀도함수 f(x)는 누적분포함수 F(x)의 미분이다.**  
E(X) = (-∞, ∞)에서 ∫ xf\_x(x) dx  
Var(X) = E(X^2) - (E(X))^2

### PDF(확률밀도함수, Probability <U>Density</U> Function)  
연속분포를 설명할 때 가장 자주 쓰는 함수  
  
**확률밀도함수는 확률을 나타내는 것이 아니다! 확률밀도를 나타내는 것이다.**  
확률밀도 → 확률을 질량이라고 생각해보자.
조약돌 세계처럼 진흙으로 이루어진 세계가 있다고 가정해보자.  
진흙의 총 질량은 1이며, 밀도라 하면 부피당 질량 or 넓이당 질량 정도가 생각날 수 있을 것이다.  
그렇게 어떤 것당 확률이며, 이는 그냥 확률이 아니다.  
  
Definition : 확률 변수 X에 대해 확률밀도함수 f(x)가 있다면,  
f(x)를 적분하면 x에 대한 확률이 나온다.
![](https://latex.codecogs.com/gif.latex?P%28a%20%5Cleq%20X%20%5Cleq%20b%29%20%3D%20%5Cint_%7Ba%7D%5E%7Bb%7Df%28x%29dx%5C%20%5C%20for%5C%20all%5C%20a%2Cb%20%5C%20%5C%20%5C%20%5Cleft%5B%20%5C%20a%20%3D%20b%20%5CRightarrow%20%5Cint_%7Ba%7D%5E%7Ba%7Df%28x%29dx%20%3D%200%20%5Cright%20%5D)  
  
즉, f(x)는 확률이 아니며, f(x)의 적분 값이 확률인 것이다.  
**확률밀도를 적분하면 확률이 나온다.**  
그리고 [] 안에서 확인할 수 있지만, 특정 값에 대한 확률은 0이 된다.   
  
확률질량함수가 유효하기 위해서는,  ![](https://latex.codecogs.com/gif.latex?f%28x%29%20%5Cgeq%200%5C%20%2C%5C%20%5Cint_%7B-%5Cinfty%7D%5E%7B%5Cinfty%7Df%28x%29dx)
(1)음수이면 안되고, (2)전체를 적분해서 1이 되어야 한다.  
  
![](https://cphinf.pstatic.net/mooc/20180831_44/1535647930742Gru0p_PNG/12-11.png)  
  
![](https://latex.codecogs.com/gif.latex?f%28x_%7B0%7D%29%5Cepsilon%20%5Capprox%20P%28X%20%5Cin%20%28x_%7B0%7D-%5Cfrac%7B%5Cepsilon%7D%7B2%7D%2C%20x_%7B0%7D&plus;%5Cfrac%7B%5Cepsilon%7D%7B2%7D%29%29%5C%20for%5C%20%5Cepsilon%3E0%2C%5C%20very%5C%20small.)  
ε이 아주 작은 수인 경우,  
정확히 x_0일 확률은 0이지만 x_0을 포함한 ε 크기의 범위에서는 확률변수 X가 들어갈 확률이 X가 정확히 x_0일 확률에 근사하고,  
그것이 확률밀도 \* 범위의 길이가 된다.  
**즉, 이것을 통해 확률밀도가 확률 차원으로 바뀌는 것이다.**  
  
ε는 아주 작으므로, 적분식의 굉장히 좁은 범위 안에서 f(x)는 상수함수가 되고,  
상수(f(x)) \* 범위의 크기(ε)가 되는 것이다.

### Variance(분산)   
분포 퍼짐의 정도 : 분포가 분포의 평균으로부터 얼마나 떨어져 있는지의 평균 
Definition:   
Var(X) = E(X-EX)인데 이 경우 선형성에 의해 E(X) - E(E(X)) = E(E(X)) = E(X)가 되어 0이 되어버린다.  
그리고 쓸모없어지므로 이 문제를 해결하기위해 절대값을 쓰면 된다.  
하지만 절대값의 함수는 V형으로 나타나고, 이는 미분을 할 수 없게 되어 다루기 힘들어진다.    
그래서 대신 전체를 제곱하여 써보자.  
  
**Var(X) = E(X-EX)^2**  
X가 평균값으로부터 얼마나 멀리 있는지의 평균인데, 그 거리의 제곱의 평균인 것이다.  
다만 제곱으로 단위가 바뀌는 점이 있다.  
  
그래서 생긴 것이  **표준편차(standard deviation)**이다.(단위를 그대로 돌려놓음)  
![](https://latex.codecogs.com/gif.latex?SD%28X%29%20%3D%20%5Csqrt%7BVar%28X%29%7D)  
  
실제로 계산할 땐 분산이 표준편차보다 이용하기 쉽다.    
  
분산을 계산하는 또 다른 방식을 알아보자.  
  
![](https://latex.codecogs.com/gif.latex?Var%28X%29%20%5C%5C%3D%20E%28X%20-%20EX%29%5E%7B2%7D%20%5C%5C%3D%20E%28X%5E%7B2%7D-2X%28EX%29%20&plus;%20%28EX%29%5E%7B2%7D%29%20%5C%5C%3D%20E%28X%5E%7B2%7D%29%20-%202E%28X%29E%28X%29%20&plus;%20%28EX%29%5E%7B2%7D%20%5C%5C%3D%20E%28X%5E%7B2%7D%29%20-%20%28EX%29%5E%7B2%7D)

### 균등분포 : Unif(a,b) 
![](https://cphinf.pstatic.net/mooc/20180831_156/1535648736073BL9K3_PNG/12-23.PNG)  
[a,b] 중점을 기준으로 나누고 완전 무작위로 뽑을 경우에는 뽑힐 확률이 앞에 반과 뒤에 반이 동일하다.  
**균등분포는 특정 범위가 뽑힐 확률이 그 범위의 크기에 비례하는 분포이다.**  
Unif : Probability ∝ length  
  
![](https://latex.codecogs.com/gif.latex?PDF%5C%20%5C%20f%28x%29%20%3D%20%5Cleft%5C%7B%5Cbegin%7Bmatrix%7D%20c%2C%5C%20if%5C%20a%5Cleq%20x%20%5Cleq%20b%20%5C%5C%200%2C%5C%20otherwise%20%5Cend%7Bmatrix%7D%5Cright.%20%5CRightarrow%201%20%3D%20%5Cint_%7Ba%7D%5E%7Bb%7Dc%5C%20dx%5C%20%5C%20%5Ctherefore%20c%20%3D%20%5Cfrac%7B1%7D%7Bb-a%7D)

기댓값은 균등분포이므로 a와 b의 중간값이다.  
  
그렇다면 이제 분산을 구해보자.  
분산을 구하려면 E(X^2)값을 알아야 한다.  
  
X^2도 확률변수이니 Y라고 치환해서 풀어보자.  
Y = X^2  
E(X^2) = E(Y) , X가 연속변수이고 E(Y)를 구하기 위해 Y의 확률밀도함수를 구하고, Y와 확률밀도함수의 곱을 적분하면 될 것이다.  
  
그렇다면 Y의 확률밀도함수가 필요할까?  
  
무의식적인 통계학자의 법칙(**L**aw **o**f **t**he **u**nconscious **S**tatistication)  
: 제대로 생각을 안하고 대충 X를 X^2로만 바꿔서 생각 없이 하게 되는 것들  
  
확률변수 X의 확률밀도 함수를 알고 X에 대한 함수의 기댓값을 구하려 할 때 생긴다.  
근데 이것이 성립을 한다!!  
  
![](https://latex.codecogs.com/gif.latex?E%28g%28X%29%29%20%3D%20%5Cint_%7B-%5Cinfty%7D%5E%7B%5Cinfty%7Dg%28x%29f_%7BX%7D%28x%29dx)  
  
다시 균등분포의 분산을 구해보자.  
  
u ~ Unif(0, 1)(0과 1사이에 위치한 균등분포 u)가 있다.     
E(u) = 1/2.  
![](https://latex.codecogs.com/gif.latex?E%28u%29%20%3D%20%5Cint_%7B0%7D%5E%7B1%7Du%5E%7B2%7Df_%7Bu%7D%28u%29du%20%3D%20%5Cint_%7B0%7D%5E%7B1%7Du%5E%7B2%7Ddu%5C%20%28%5Cbecause%20f_%7Bu%7D%28u%29%20%3D%201%29%20%3D%20%5Cfrac%7B1%7D%7B3%7D)  
  
**Var(u) = E(u^2) - (Eu)^2 = 1/3 - 1/4 = 1/12**  
  
확률밀도함수가 상수이기 때문에 연속분포 중에 균등분포가 가장 쉬워보인다.  
균등분포의 범위는 제한된 범위가 필요하다.  
모든 실수를 범위로 두면 정규화를 할 수 없다.(적분해서 1이되는 상수를 찾을 수 없어짐.)    
0에서 1까지의 균등분포는 아주 특별하다.  
만약 확률변수로 다른 분포가 궁금할 경우 모의로 그 분포를 만들 수 있다.  
  
이를 **'균등분포의 일반성(Uniform is Universal)'**이라고 한다.  
  
균등분포가 주어졌을 때 어느 분포든 만들 수 있다.  
  
u ~ unif(0,1)일 때, F가 우리가 원하는 누적분포함수(CDF)라고 하자.  
(단, F는 가파르게 증가하며, 연속함수이다. → F의 역함수 존재)  
우리는 특정 F를 가진 분포를 만들 고 싶다.  
  
X = F^(-1)(u). 그리고 X ~ F(X가 F의 분포를 따른다).  
  
Proof ) P(X ≤ x) = P (F^(-1)(u) ≤ x) =  P(u ≤ F(x)) = F(x)  
![](https://cphinf.pstatic.net/mooc/20180831_6/1535650689713v34iq_PNG/12-3.PNG)  
Unif(0,1)에서 x까지의 확률질량은 0에서 x까지 선분의 길이와 같다.
