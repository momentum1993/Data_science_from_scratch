### 포아송분포 (The Poisson distribution)  (이산형 분포)
  
PMF : ![](https://latex.codecogs.com/gif.latex?P%28X%3Dk%29%20%3D%20%5Cfrac%7Be%5E%7B-%5Clambda%7D%5Clambda%5E%7Bk%7D%7D%7Bk%21%7D%2C%20k%20%5Cin%20%5C%7B0%2C1%2C2%2C...%5C%7D)()  
이항분포에서는 0~n까지 k의 범위가 정해져 있지만, 포아송분포의 경우 음수가 아닌 어느 정수든 와도 상관없다.  
그리고 λ는 **비율 모수(rate parameter)**라고 한다.(단, λ는 양의 실수인 상수)  
  
유효성 검사 : 다 더해서 1이 되나?

기댓값 계산: X ~ Pois(λ)(확률변수 X가 포아송 분포를 따른다)라고 하자.

포아송분포는 이산형 분포이며, 많이 쓰이고 있다.  
수를 세는 응용문제에서 많이 쓰이는데,  
많은 시도에 따라 각 결과가 성공과 실패가 나온다고 할 때의 성공 횟수 #을 센다고 하자.  
각 시도의 성공확률은 매우 낮으며, 많은 시도가 뒷받침 될 때가 일반화한 응용의 조건이다.

### 포아송 패러다임(poisson paradigm)(=포아송 근사,poisson approximation)
어떤 큰 숫자 n에 대하여  A_1, ..., A_n의 사건들이 각각  P(A_j) = p_j 라는 낮은 확률로 발생하고,  
각 사건은 독립이거나 weakly dependent(약한 의존성, 아주 낮은 확실성으로 유추할 수도 있을 때 )일 때,   
발생하는 사건(A_j)의 수는 Pois(λ)의 분포를 따른다.  
  
그리고 전에 계산한 포아송분포의 기대값은 λ였으모로, λ = ∑ P_j가 된다.  
  
이항분포(n, p)가 (n이 커지고, pp가 작아질 때) 포아송 분포로 수렴하는 것을 증명할 것이다.  
  
(포아송분포의 경우 확률(p)가 달라도 되고, 조금의 의존성을 허용한다.)  
  
X ~ Bin(n,p)이고, n→∞, p→0, λ = np(λ는 상수)라고 하자.(이항분포의 기댓값 np와 포아송 분포의 기댓값 λ를 이어줌으로써 두 분포를 연결, 즉 n과 p가 극한으로 가는 속도가 같다는 것)  
이 때 확률질량함수 P(X=k)는 어떻게 될지 보아라.  
  
![](https://latex.codecogs.com/gif.latex?P%28X%20%3D%20k%29%20%3D%20%5Cbinom%7Bn%7D%7Bk%7Dp%5E%7Bk%7D%281-p%29%5E%7B%28n-k%29%7D%2C%5C%20k%5C%20fixed%5C%5C%20%3D%20%5Cfrac%7Bn%28n-1%29...%28n-k&plus;1%29%20%5Clambda%5E%7Bk%7D%7D%7Bk%21n%5E%7Bk%7D%7D%281-%5Cfrac%7B%5Clambda%7D%7Bn%7D%29%5E%7Bn%7D%281-%5Cfrac%7B%5Clambda%7D%7Bn%7D%29%5E%7B-k%7D%28%5Cbecause%20p%20%3D%20%5Cfrac%7B%5Clambda%7D%7Bn%7D%29%5C%5C%20%3D%20%5Cfrac%7B%5Clambda%5E%7Bk%7D%7D%7Bk%21%7De%5E%7B-k%7D%20%28%3D%20Pois%5C%20PMF%5C%20when%5C%20X%3Dk%29)  
  
n(n-1)...(n-k+1)는 n개이고, n^k 또한 n개 이다.  
그래서 n/n, n-1/n, ... , n-k+1/n와 같이 n개로 나눌 수 있는데 n→∞일 때 각 값은 1로 수렴한다. 그래서 1을 n개 곱하면 1이 된다.  
그리고 (1 - λ/n)^(-k) 에서 n→∞이 되면 1-0^k으로 1이 되게 된다.
(1 - λ/n)^n 은 n→∞일 때 (1+x/n)^n → e^x가 된다는 것을 이용하면, e^(-λ)가 된다.  
  
ex1) 길바닥에 빗방울이 떨어지는 횟수를 예시로 들어보자.  
큰 사각형으로 보면 빗방울이 거의 떨어지겠지만, 작은 사각형의 경우 빗방울이 떨어질 확률이 매우 낮을 것이다. 그리고 여기서 λ는 비가 얼마나 세게 오는지 이다.  

이 상황에서 모든 사각형이 맞을 빗방울이 독립적이고, 각 사각형마다 빗방울을 맞을 확률이 p로 동일하다면 이항분포를 그대로 따를 것이다.  
그리고 이항분포라면 0,1의 값을 가질 수 있는데 2개의 빗방울이 한 사각형에 들어갈 수 있을 것이다. 이항분포가 맞다고 해도 매우 큰 n과 작은 p로 인해 계산하기 힘들어진다.(∵ 팩토리얼 계산)

![](https://cphinf.pstatic.net/mooc/20180830_25/1535612742224Ly3WS_PNG/11-1.PNG)  


그래서 길바닥에 빗방울이 떨어지는 횟수를 간단하게 포아송 근사로 설명할 수 있다. 각 사각형에 빗방울이 떨어지는 사건은 이항분포이지만, 그 사건은 서로 독립이다. 빗방울은 많이 떨어지지만 한 사각형 안에 떨어질 확률은 작기 때문에, 포아송 분포로도 볼 수 있다.  
  
  
ex2) n명 중에서 생일이 같은 세 명의 사람을 찾을 확률의 근사값(approximation)?  
  
먼저 포아송 분포 근사가 이 문제에 적절한가?
  
일단 n이 크다고 가정하자.  
n C 3의 3명 집단을 만드는데, 사람들이 1 ~ n까지 표시되어 있으며, 그 중에서 3명으로 이루어진 그룹을 만든다. 그 후 각 그룹당 3명의 생일이 같은지 물어본다.  
-> 각 집단의 지시확률변수 I_ijk(i,j,k는 각각 사람의 라벨이며, i\<j\<k)  
  
E(3명의 생일이 같음) = n C 3 ㆍ 1 / 365^2  
  
X = (3명이 생일이 같을 경우의 수)  일 때,
Pois(λ)의 근사, λ = n C 3 ㆍ 1 / 365^2 (n은 충분이 크다.)  
각 시도의 확률은 매우 낮으며, 사건은 독립적이지 않다.  
(I_123과 I_124는 독립 X 이며, I_123이 1이면 뒤의 I_124에서 1과2는 생일이 같다는 것을 알고 있으므로 I_124는 1일 확률이 높아진다. = 약한 의존성(weak dependency) )  
  
적어도 한 집단이 생일이 같을 확률  
P(X ≥ 1) = 1 - P(X = 0) ≒ 1 - e^(-λㆍλ^0/0!) = 1 - e^(-λ)  
  
이렇게 간단하게 포아송분포 근사로 알아낼 수 있다.
