# 13강- 정규분포 (Normal Distribution)

### 균등분포의 일반성

0에서 1까지의 균등분포는 아주 특별하다.  
만약 확률변수로 다른 분포가 궁금할 경우 모의로 그 분포를 만들 수 있다.  
  
이를 **'균등분포의 일반성(Universality of Uniform)'**이라고 한다.  
  
균등분포가 주어졌을 때 어느 분포든 만들 수 있다.  
  
u ~ unif(0,1)일 때, F가 우리가 원하는 누적분포함수(CDF)라고 하자.  
(단, F는 가파르게 증가하며, 연속함수이다. → F의 역함수 존재)  
가파르게 증가시키는 이유는 평평한 부분을 없애기 위해서(단, 원래 누적분포함수는 평평한 부분 지닐 수 있다.)  
일반적으로 우연속만 만족하면 된다.  
  
우리는 특정 F를 가진 분포를 만들 고 싶다.  
  
X = F^(-1)(u). 그리고 X ~ F(X가 F의 분포를 따른다).  

이때, 균등분포를 가지는 확률변수로 우리가 원하는 분포의 형태를 가지는 확률변수로 시뮬레이션이 가능하다.  
  
대부분 F의 역함수를 찾는게 어렵겠지만, 이론적으로는 시뮬레이션 가능하다.  
  
Proof ) P(X ≤ x) = P (F^(-1)(u) ≤ x) =  P(u ≤ F(x)) = F(x)  
![](https://cphinf.pstatic.net/mooc/20180831_6/1535650689713v34iq_PNG/12-3.PNG)  
Unif(0,1)에서 x까지의 확률질량은 0에서 x까지 선분의 길이와 같다.  

##### 이러한 균등분포의 일반성의 예시들을 보자.
  
ex1) X에서 균등분포로 가는 법  
0부터 1까의 균등분포로 시작해서 F^(-1)(u)를 계산한 것이 누적분포함수라고 했다.
X를 이미 알고 있다면??  
X ~ F일 때, F(X) ~ Unif(0,1) 
  
누적분포함수 F(x)에 대해서,  
F(X) = P(X ≤ x) 에 x에 X를 대입할 경우,  
F(X) = P(X ≤ X) = 1 → 항상 참이 되어버리고 이는 잘못된 것이다.  
  
F를 식으로 풀어 쓴 후 x에 X를 대입할 경우,  
F(x) = 1 - e^(-x), x>0 (x가 양수일 때 이 함수는 연속이고 강한 증가, 음수는 0)  
F(X) = 1 - e^(-X)  

문제가 없다. 이렇게 F를 식으로 풀어 쓴 후 바꾸면 된다.  
  
X가 아주 복잡하거나 모르는 분포를 지닐 때, 모델을 실험하고 싶을 때 간단한 균등분포로 바꿀 수 있다는 것이다.  
  
ex2) 모의 실험(시뮬레이션)의 예시
  
F(x) = 1 - e^(-x), x>0 (Expo(1))이고, u~Unif(0,1)일 때,  
F(x)의 분포를 시뮬레이션하고 싶다. (X ~ F)  
  
F^(-1)(u) = -ln(1-u) ⇒ F^(-1)(U) = -ln(<U>1-U</U>) ~ F  
  
여기서 1-u에 대해서 말하고 싶은데,  
  
1-u ~ Unif(0,1) **(균등분포의 대칭성, Symmetry of Unif)** : u는 0~u까지의 거리이고, 1-u는 u~1까지의 거리며, 대칭이 존재하며, 왼쪽에서 재든 오른쪽에서 재든 이 둘은 모두 균등분포를 따른다.  
a+ bu 도 어떠한 구간에서 균등분포일 것이다. **(균등분포의 선형변환)** 
예를들어 0~1까지의 균등분포로 시작해서 10을 곱하면 0~10의 균등분포가 된다.  
(다만 균등분포를 따르는 u의 선형변환에 대해서만 보장. )  
비선형 변환(u의 n차식(n ≠ 1인 실수))은 대부분 균등분포를 따르지 않을 것이다.  
예를 들어 u^2의 경우 (0,1)을 만족한다고 균등분포임을 확신 할 수 없다. → 누적분포함수를 확인해야함.  
  
  



###  확률변수 간의 독립성    
  
Definition :   
    X_1, ... , X_n 까지의 확률변수가 있을 때,  
    모든 x_1, x_2, ..., x_n에 대하여  
    <연속분포(continuous) 의 경우>
    P(X_1 ≤ x_1, ... , X_n ≤ x_n)(결합 누적분포함수) = P(X_1 ≤ x_1)ㆍ ... ㆍP(X_n ≤ x_n)  
    <이산분포(discrete)의 경우>  
    P(X_1 = x_1, ... , X_n = x_n)(결합 확률질량함수) = P(X_1 = x_1)ㆍ ... ㆍP(X_n = x_n)  
    를 만족할 경우 확률변수들이 독립적이다.  
      
위처럼 간단하게 정의내릴 수 있지만, 실제로는 많은 식을 적은 것이다.  
한 개의 식처럼 보이지만, 모든 값에 적용돼야 하기 때문.  
  
완전한 독립성은 이 중에서 어떠한 조합의 사건 실행 여부를 알아도, 나머지에 대해 아무 것도 모르는 것이다.    
  
  
ex) 쌍으로 독립은 되지만, 완전 독립성은 지니지 않는 예제를 한 번 보자.  
  
X_1, X_2 ~ Bern(1/2) (동전을 두 번 던진다고 해보자.) 그리고 이는 i.i.d일 때,  
X3 = 0(if X_1 = X_2), 1(otherwise) 라고 하자.  
  
이 둘은 쌍(pairwise)으로 독립이지만, 완전한 독립은 아니다.  
  
X_1과 X_2를 알면, X_3을 알 수 있다.  
(X_1과 X_3는 독립, X_2와 X_3는 독립, X_1과 X_2는 독립)  

### 정규 분포(Normal Distribution)  
  
(중심극한정리(central limit theorem): 여러 개의 독립적이고 동일한 확률변수를 많이 더하면 정규분포에 근사한다.) → 정규분포가 근본적인 이유 중 하나
