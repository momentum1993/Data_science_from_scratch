# 6강- Monty Hall 문제와 심슨의 역설 (Monty Hall, Simpson's Paradox)

### Monty Hall 문제 

몬티 홀은 게임 쇼 진행자이다.  
진행자가 게임 쇼에서 진행한 게임에서 나온 것.  
몬티 홀은 똑같은 3개의 문들 중 1개의 문을 고르라고 하는데  
각각의 문들 중에서 1개의 문 뒤에는 자동차, 그리고 나머지 2개에는 염소가 있다.  
그리고 **'몬티 홀만 어떤 문 뒤에 자동차가 있는지 알 수 있다'**
![](https://cphinf.pstatic.net/mooc/20180822_5/15348681563919FjI3_PNG/6-1.png)  
우리가 처음에 1번 문을 골랐다고 해보자.  
그렇다면 몬티는 2, 3번 중에 염소가 있는 2번 문을 열고, 다시 1,3번 중에 다시 고를 기회를 준다.  
그리고 몬티가 어떤 문을 열지 선택할 수 있다면, 그는 동일한 확률로 결정을 내릴 것이다.  (처음 우리가 선택한 문 뒤에 차가 있을 경우 = 선택이 옳았을 경우에만 적용된다)  
여기서 우리는 처음 선택한 1번을 바꿔야 하는가?  
  
일단 답은 바꿔야 한다이다.  
바꾸지 않고 성공할 확률 1/3, 바꿀 경우 성공할 확률 2/3 이기 때문.  
(2개 문 중 1를 고르는 것이기 때문에 1:1로 생각할 수 있지만 , 우리는 1번을 고르면서 얻은 증거가 있기 때문에 확률이 달라짐)  
  
몬티가 2번을 연다고 가정을 하자.(= 2번 문 뒤에 염소가 있다는건 확실한 사실)  
조건(condition)을 걸 때는 증거에 입각해야 한다.  
**-> 몬티가 2번 문을 열었다는 사실 또한 중요하다 !!**  

(1) Tree Diagram  
먼저 우리가 1번을 고르면  
몬티는 (1) 어떤 문에 자동차가 있는지 (2) 어떤 문을 열것인지 고려해야 한다. 
![](https://cphinf.pstatic.net/mooc/20180822_20/1534868342212siz3w_PNG/6-2.png)  
  
(2) conditional probability  
**전체 확률의 법칙**을 쓸 때 중요한 것이 어떤 걸 조건으로 삼을지 정하는 것.  
  
우리는 자동차가 어느 문 뒤에 있는지 알고 싶어한다.  그렇기에 이것을 조건으로 삼을 것.  
  
S : 우리가 성공하는 사건(처음 선택을 바꾸는 전략을 항상 사용한다는 가정 하에)  
D_j : J번 문이 차를 갖고있는 사건.

P(S) = P(S | D_1)ㆍ1/3 + P(S | D_2)ㆍ1/3 + P(S | D_3)ㆍ1/3  
     = 0 + 1ㆍ1/3 + 1ㆍ1/3  
     (-> P(S | D_1)ㆍ1/3 = 0 인 이유는 처음에 1번 선택후 다른 선택을 할 것이기 때문)  
     = **2/3**  
     
여기서 비조건부확률일 때만 이 전략이 통한다는 것을 인지해야 한다.  
  
Symmetry(대칭성)에 의해서, P(S | 몬티가 2번문 엶) = 2/3 (이것은 몬티홀이 2번 혹은 3번 문을 열었을 경우에만 해당)  
  
문을 열지 않았을 때 몬티가 각 문을 열 확률이 동일한 대칭성을 띄기에 이러한 답이 나온다.  
그렇지 않다면 조건부 확률은 비대칭성 때문에 달라질 것.  

### Simpson's Paradox(심슨의 역설)  
역설은 존재하지 않는다. 다만, 사람들의 직관과 반대될 뿐.  
  
A,B 2명의 의사가 있다고 하자.
(1) A는 여러 종류의 수술집도에 있어 모두 B보다 성공률이 높다.(각각의 항목의 성공률)
(2) B는 모든 수술에 대한 집도 성공률이 A보다 높다.(전체 성공률)  

그리고 심슨의 역설은 이것이 가능하다고 하는 것이다.  

여기 히버트, 닉 2명의 의사와 2가지의 병에 대한 수술이 있다고 가정해보자.  
  
![](https://cphinf.pstatic.net/mooc/20180828_196/1535385173902NcLqR_PNG/6-3.png)  
  

비조건부인 경우 닉 > 히버트이다.  
그러나 심장수술, 반창고 제거 등의 조건을 준다면 히버트 > 닉 이다.  
  
어떤 사람이 수술을 받는다고 가정해보자.  
A : 수술이 성공하는 사건  
B : 의사 닉이 수술하는 사건 (B^c = 의사 히버트가 수술하는 사건)  
C : 심장 수술하는 사건  (C^c = 반창고 제거하는 사건)  
  
P(A | B, C) < P(A | B^c, C)  
P(A | B, C^c) < P(A | B^c, C^c)  
P(A | B) > P(A | B^c)  
  
C가 교란요인(confounder)이다.  
C라는 요인을 조건으로 삼아 조건부확률을 보지 않는다면 오해할만한 답을 얻게 될 것이다.  
  
심슨의 역설이 틀렸음을 증명할 수 있는가?  
  
전체 확률의 법칙을 통해 보일 것.(P(A | B)를 P(A | B, C)와 연관 지을 것임)  
  
P(A | B) = <U>P(A | B, C)</U>ㆍP(C | B) + <U>P(A | B, C^c)</U>ㆍP(C^c | B)  
( P(A | B, C) < P(A | B^c, C) , P(A | B, C^c) < P(A | B^c, C^c) )   
대소 비교가 가능한 밑줄친 부분 이외에 P(C | B)와  P(C^c | B)를 가중치로 뒀을 경우  
P(C | B), P(C | B^c)와  P(C^c | B), P(C^c | B^c)는 서로 대소 비교 불가.  
가중치가 변하기 때문에 심슨의 역설이 일어날 수 있는 것이다.
