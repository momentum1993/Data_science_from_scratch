# 9강- 기댓값, 지시확률변수와 선형성 (Expectation, Indicator Random Variables, Linearity)

누적분포함수(CDF)는 모든 확률변수에 대해서 성립한다.  

확률변수 X가 있고, F(x) = P(X ≤ x) , x는 실수를 만족하는 함수.  
X가 오로지 정수 값을 받는 것이라도 해도, 이 함수는 모든 실수 X에 대해 적용 가능하다.  
  
이산(discrete) 확률 변수가 있을 때
![](https://cphinf.pstatic.net/mooc/20180829_225/1535548770078JbR96_PNG/9-1.png)  
이러한 모양으로 누적분포함수의 값이 나온다.  
만약 X가 가질 수 있는 값이 무한이라면 값이 증가할수록 점프는 계속 하겠지만 1에 도달할 수 는 없다.  
  
그리고 여기서 각 정수에서 점프하는 높이는 X = 각 정수일 확률이고,  
여기서 **점프의 높이가 확률질량함수(PMF)가 된다.**
  
확률질량함수를 더해서 누적분포함수를 알 수 있으며, 누적분포함수에서 점프의 높이를 통해 확률질량함수를 알 수 있다.  
  
P(1 ＜ X ≤3 )를 CDF의 관점에서 찾아보자. (**누적분포함수는 모든 분포를 알려주기 때문에 여기서 X에 대한 모든 확률을 찾아볼 수 있을 것**)

P(X ≤ 1) + P(1＜ X ≤3) = P(X ≤ 3)  (확률의 공리에 의해 사실)
 => P(1＜ X ≤3) = F(3) - F(1)  
 
**이를 통해 P(a＜ X ≤b) = F(b) - F(a) 임을 알 수 있다.** 
  
#### 누적분포함수의 3가지 속성(Properties)  
(1) 오직 증가한다(increasing)  
(2) 우연속함수(right continuous)(= 함수의 어떠한 점을 잡고 오른쪽에서 접근해 극한을 찾으면 이 함수의 값은 이 점의 값으로 수렴한다.).  
(3) X→ −∞ 으로 갈수록 F(X)→0, X→∞으로 갈수록 F(X)→1.  
  
이 세가지를 만족하는 함수는 유효한 누적분포함수이다.(필요충분조건)    
  
  

### 확률변수의 Independence(독립성)  
  
X,Y는 둘다 독립인 확률변수이다. 

P(X ≤ x, Y ≤ y) = P(X ≤ x)ㆍP(Y ≤ y) (모든 x,y에 대해서)
(! 독립성은 곱하는 것. P(X ≤ x, Y ≤ y)와 같은 함수를 '결합누적분포함수'라고 한다.)  
  
이산확률변수의 경우 더 확률질량함수를 더 다루기 쉽다.  
이산확률변수의 결합확률질량함수 
P(X=x, Y=y) = P(X=x)ㆍP(Y=y)  

하지만 위의 정의는 연속적인 경우에는 성립하지 않음.  
  
### 평균(averages)  =  Means(평균값), Expected Values(기대값)   
  
확률변수는 직접 실험하기 전에는 모른다.  
그러나 이전에 평균으로 예측해 볼 수 있다.  

(1) 다 더해서 나누기  
1,2,3,4,5,6 -> 1+2+3+4+5+6/ 6  = 3.5  
위와 같은 등차수열의 경우 첫 값인 1과 끝 값인 6을 2로 나누어도 값이 같음.    
  
(2) 가중 평균(weighted average)  
1,1,1,1,1,3,3,5 -> 5/8 \*1 + 2/8\*3 + 1/8\*5   
  
그리고 이를 확률변수에 적용해보자.  
  
이산확률변수의 평균 X  
E(X)(=기대값) =  ∑xㆍP(X=x)(P(X=x)>0)  
  
X ~ Bern(P)  
E(X) = 1ㆍP(X=1) + 0ㆍP(X=0) = P  
X = 1,0  (지시확률변수 : 어떤 사건이 일어나면 1, 일어나지 않으면 0)  
그래서 1일 확률은 A사건이 일어날 확률과 동일하다.  
**E(X) = P(A)**  (fundamental bridge = 기대값과 확률을 이어줌)
  
X ~ Bin(n,p)

### Linearity(선형성)  
  
E(X+Y) = E(X) + E(Y)  (이는 X와 Y가 의존적이어도 성립. 항상 성립한다.)  
  
E(cX) = cE(X) (c는 상수)  
  
그렇다면 X ~ Bin(n,p) 의 이항식을 다시 한번 해보자.  
  
Bin(n,p)는 각각의 Bern(p)가 가지는 기대값 p가 총 n개가 있으므로 np가 된다( 선형성에 의해 )  

X = X_1 + X_2 + ... + X_n , X_j ~ Bern(n,p) (X_j는 각각의 베르누이 시행)  
E(X) = nㆍE(X_1) = np  
  
ex) 52장의 카드에서 5카드를 뽑는데, X = 에이스의 개수.  
그리고 X_j를 j번째 카드가 에이스일 지시확률변수라고 하자.(일단 순서가 있다고 생각하고) -> 1 ≤ j ≤ 5    
  
E(X)  
= E(X_1 + X_2 + .. + X_5)(지시확률변수에 의해)  
= E(X_1) + ... + E(X_5)(선형성에 의해)  
= 5ㆍE(X_1)(대칭성에 의해)  
= 5ㆍP(첫번째 카드가 에이스일 확률)(fundamental bridege -> 기대값과 확률을 이어주는 다리)  
= 4/52 = 5/13  
(X_j가 의존적이지만 이러한 선형성이 그대로 작용한다.)  
어떠한 초기하(Hypergeometric)의 기댓값을 준다.  
  
### 기하분포(geometric distribution)  
  
Geom(p) : 이항식처럼 독립적인 베르누이 시행이 있다고 가정한다. 그리고 같은 시행을 여러번 반복한다. 그래서 Bern(p)는 동일한 성공 확률을 가진다.  
기하는 첫 성공 전에 얼마나 많은 실패가 나오는지의 횟수 #을 알려준다.  
X~Geom(p) 라고 할 때(이는 n번의 횟수가 정해진 이항분포와는 다르게 성공하기 이전까지 계속되며 그 시점까지의 실패 횟수를 센다.)  
q = 1-p  
  
FFFFFS  
P(X=5) = q^5 ㆍ p
  
PMF(확률질량함수) : P(X = k) =  q^kㆍp, k ∈ {0,1,2,...}
