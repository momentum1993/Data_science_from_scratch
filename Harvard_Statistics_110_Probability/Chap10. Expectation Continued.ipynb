### Linearity(선형성) 증명

T = X+Y라 할 때,  
E(T) = E(X) + E(Y)  

기댓값 E(T)를 이산확률변수에 대해서 증명을 해보자면
일단 기댓값 E(T)는 ∑t X 확률질량함수 꼴(=가중평균)로 나타낼 수 있음을 지난 시간에 배웠다.

그리고 우리는 P(T=t)를 가지고 X=x를 알고있는 '조건화' 전략으로 한번 풀어보자.  
![](https://latex.codecogs.com/gif.latex?%5Csum_t%20tP%28T%20%3D%20t%29%20%3D%20%5Csum_t%20tP%28T%20%3D%20t%7CX%20%3D%20x%29P%28X%20%3D%20x%29)

X와 Y가 독립적이라는 조건이 없다.  
  
그래서 다른 관점으로 한 번 보자(조약돌세계)

유사하게, **E(cX) = cE(X) (c는 상수)**  

위의 식은 종속적인 경우도 E(X+Y) = E(X) + E(Y)가 성립되기에 만들어지는데 그렇다면  
종속의 극한적인 경우를 보자 : X=Y일 때의 경우
E(X+Y) = E(2X) = 2E(X) = E(X) + E(Y)   
  
ex) 음이항분포(Negative Binomial)

#### 첫 성공분포(First Success Distribution)  
  
X∼FS(p)(첫 성공까지 걸리는 시간이며, 이는 성공횟수를 센다.)  
  
Y = X - 1이라고 두자. 그리고 Y ~ Geom(P)  
E(X) = E(Y+1) = E(Y) + 1 = q/p + 1 = **1/p** 
(성공 확률이 1/10이라면 평균적으로 첫 성공을 하려면 10번의 시행을 거쳐야한다는 것)

#### 기댓값 ex) 1~n까지의 랜덤한 순열이 있다.(모든 순열의 확률은 같고, n ≥ 2)  
그리고 극댓값(local maxima)의 평균 도는 기댓값을 찾아라.  
  
**3** 2 1 4 **7** 5 **6**  
3,6,7은 극댓값이며, 주변 이웃 숫자보다 큰 숫자.  
(극댓값이 나오는 경우는 제일 처음과 끝, 그리고 가운데에 있을 수 있다.)  
  
I_j를 극댓값을 지닌 자리 위치 j에 대한 극댓값의 지시확률변수(indicator random variable)이라고 가정하자.(j는 1~n)  
  
이제 I_j에 대한 합은 극댓값의 개수가 된다.  
  
![](https://latex.codecogs.com/gif.latex?E%28I_%7B1%7D&plus;...&plus;I_%7Bn%7D%29%20%3D%20E%28I_%7B1%7D%29&plus;...&plus;E%28I_%7Bn%7D%29%20%3D%20%5Cfrac%7Bn-2%7D%7B3%7D&plus;%20%5Cfrac%7B2%7D%7B2%7D%3D%20%5Cfrac%7Bn&plus;1%7D%7B3%7D)  
n-2개의 중간의 숫자들 중에 3개의 숫자를 뽑아낸다.  
그리고 3개의 숫자의 배열은 6가지가 있고 가장 큰 숫자가 가운데에 있는 방법은 2가지이다.  
그래서 가장 큰 숫자가 가운데 있을 확률은 각각 1/3이 된다.  
(중간 숫자와 끝 숫자를 구별할 수 있다면 대칭성을 이용할 수 있다.)  
그래서 n-2 / 3  
그리고 끝의 2개의 숫자에 대해서 어떤 숫자가 큰 숫자일 확률은 1/2이므로  
2/2가 된다.  
  
→ 간단한 상황:  n = 2 

     [[ 1 2 ]]    [[ 2 1 ]]        →  E(X)=1 

→ 극단적인 상황:  n→∞ 

극대값 수의 기댓값  E(X)→∞

#### ex2) 상트페테부르크 역설(St. Peterburg Paradox)  
  
X는 처음 동전 앞면이 나올 때(H)까지 동전을 던지는 횟수이며, x번 던지면 2^x 달러를 얻을 것이다. (=이것은 기하분포+1인 것이다. 성공횟수도 1개 포함하기 때문에)  
그리고 이 게임을 하기 위해 얼마를 내야 할까(합당한 가격일까)?  
  
Y = 2^x , 그리고 E(Y)를 찾아보자.  
  
![](https://latex.codecogs.com/gif.latex?E%28Y%29%3D%20%5Csum_%7Bk%3D1%7D%5E%7B%5Cinfty%7D2%5E%7Bk%7D%5Cfrac%7B1%7D%7B2%5E%7Bk%7D%7D%20%3D%20%5Csum_%7Bk%3D1%7D%5E%7B%5Cinfty%7D1%20%3D%20%5Cinfty)  
  
그렇다면 이번엔 수령받을 수 있는 돈의 상한선을 2^40(약 1조)로 둔다고 하자.  
  
![](https://latex.codecogs.com/gif.latex?E%28Y%29%3D%20%5Csum_%7Bk%3D1%7D%5E%7B40%7D2%5E%7Bk%7D%5Cfrac%7B1%7D%7B2%5E%7Bk%7D%7D%20%3D%20%5Csum_%7Bk%3D1%7D%5E%7B40%7D1%20%3D%2040)

그렇다면 무한이 아닌 40달러가 된다.  
  
![](https://latex.codecogs.com/gif.latex?%5Cinfty%20%3D%20E%28Y%29%20%3D%20E%282%5E%7Bk%7D%29%20%5Cneq%202%5E%7BE%28X%29%7D%20%3D%204)
2^k의 기댓값은 무한이다.  
여기서 주의할 점은 E를 움직여서 E를 지수로 올리는 것이다.    
하지만 이 경우 2^E(X) 에서 첫 성공의 경우 기댓값은 1/p이며, 여기서 p는 1/2이므로 2^E(X) = 4가 되고, 이는 무한대인  2^k의 기대값과 다르다.  
**물론 선형성을 이용하면 되지만, E를 마음대로 움직이면 안된다.**
