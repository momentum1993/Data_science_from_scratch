# 8강- 확률변수와 확률분포 (Random Variables and Their Distributions) 

이항 분포(binomial distribution)  = Bin(n,p)  
n : 임의의 양의 정수  p : 0과 1사이의 임의의 실수  
따라서 어떤 n, p에 대해서도 이항분포가 존재하게 된다.
n개의 독립적인 시행이 있고 각 시행의 결과가 성공(1) 또는 실패(0)일 때 이항분포는 성공한 횟수가 된다.  
X ~ Bin(n,p) 라고 표기하며 이는 X가 이항분포를 가지는 확률변수라는 것이다.  
  
확률분포를 해석할 때 3가지 관점으로 해석할 수 있다.  
  
1. story : X는 n개의 독립적인 시행에서 성공한 횟수 #를 뜻하며, 각 시행(= Bern(p))은 **독립적(independent)**이어야 한다.(여기서 Bern(p)는 매개변수 p의 베르누이 시행이며, 이는 성공 또는 실패 결과를 p의 확률로 갖는 시행)   
  
2. 지시확률변수(indicator random variables)의 합의 꼴로 해석한다.
 : X_n = X_1 + X_2 + X_3 + ... + X_n , X_j = 0(if j번째 시행 실패), 1 (if j번째 시행이 성공)  
 이는 복잡한 분포를 0과 1같이 단순한 수의 합으로 만든 것에 의의를 둘 수 있다.  
  X_1 + X_2 + X_3 + ... + X_n  i.i.d.(=independent identically distributed, 독립적으로 동일하게 분포되었다는 것을 의미) Bern(p)  
 각 시행은 독립적이며, 이 확률변수들이 각 시행들의 지표라고 가정했기 때문에 각 확률변수는 독립이다.  
 즉, 각 X가 같은 분포를 지닌다는 뜻이다.  
 
3. PMF(확률질량함수, Probability Mass Function) P(X = k) =  n C k ㆍ p^k ㆍ q^(n-k) , q = 1- p  , n ∈ {0,1,2,3,...,integer}
 
 Random Variables  
 어떠한 표본공간(Samplespace)에 조약돌들이 있다고 하자.  
 확률변수는 각 조약돌에 대해서 숫자를 배정하는 것이 함수이다.  
 발생 가능한 결과로 이루어진 추상적인 공간에서 시작해서 각 결과에 대해 숫자값을 지정한 것이 확률변수이다.  
 ex) X = 7 이라는 것은 사건(event)이며, 이는 표본공간의 부분집합을 뜻한다.  
 사건을 표현하는 표기법이다.
 
 CDF(누적분포함수, Cumulative Distribution Function)
 X = 7을 포함한 임의의 x에 대해 X = x는 사건이다.  
 또한,  X ≤ x 도 사건이다.  
 F(x) = P(X ≤ x)라고 할 때,  
 함수 F를 X의 누적분포함수라고 합니다.  
 
 X를 임의의 확률변수라고 두자.  
 X ≤ x는 사건이지만, 시행을 하기 전에는 X의 값이 무엇인지 알 수 없다.
 예를 들어 x가 9이고, X가 7이라는 것을 시행을 통해 알았다면 이 사건이 일어난 것이다.    
 그래서 F(x)는 사건 X ≤ x 가 일어날 확률을 뜻하는 누적분포함수이다.  
 이 함수는 분포를 설명하는 여러 가지 방법 중 한 가지의 방법이며, X에 대한 모든 가능한 확률을 결정할 수 있는 방법이다. 이는 모든 확률 변수에 대해서 이용 가능하다.   
   
 PMF는 오직 이산확률변수(discrete random variable)에 대해서만 정의된다.  
 
 이산확률변수(discrete random variable)  
 여기서 이산적이라는 말은 정수값을 지닌다는 것으로, 대표적으로 이항변수가 있다.  
 하지만 가능한 값들이 정수가 꼭 되어야 할 필요는 없지만, 열거될 수 있어야 한다.  
 유한 배열이나 무한 배열처럼..  
 a_1, a_2, a_3, ...  
   
 연속확률변수(continuous random vairable)  
 임의의 실수 값을 가지거나 특정 구간 사이의 실수값을 지닌다.(더 설명이 필요하겠지만)  
 어떤 확률변수는 이산, 연속 둘 다 아닐 수도 있으며, 두 개념을 모두 포함할 수도 있다.  
 PMF(확률질량함수)는 모든 j에 대해 X = a_j일 확률이다.  
 a_1 ~ a_n이 1부터 n까지의 정수 일 때, X = 1, X = 2 ,... 때의 확률들을 알아야 한다.  
 그 모든 확률을 명확히 정의해햐 한다.  
 **P_j = P(X = a_j)라고 했을 때, P_j ≥ 0, ∑ P_j = 1 이어야 한다.**  
 
 다시 이항 분포로 돌아와서
 P(X = k) =  n C k ㆍ p^k ㆍ q^(n-k) , q = 1- p  , n ∈ {0,1,2,3,...,integer}

X ~ Binomial(n, p) 이고, Y ~ Binomial(m, p)이고 서로 독립적일 때,
X+Y ~ Binomial(n+m, p)이다.  
  
이는 X와 Y 두 함수의 합으로 이를 위해서는 두 함수가 같은 정의역을 지녀야 하며, 여기서 정의역은 모두 조약돌 표본집단이기에 만족한다. 그래서 두 함수 값을 계산하고 그 값을 더하면 새로운 함수가 만들어 지는 것이다.  
  
1. 처음 n개 시행과 다음 m개의 시행은 서로 독립적이기에 총 횟수를 더하면 n+m번 시행을 한것이고, 두개 모두 p값이 같기에 성공한 횟수는 X+Y가 되는 것이다.  

2. X = X_1 + X_2 + ... + X_n, Y = Y_1 + Y_2 + ... + Y_n라 하면  
 모든 X_j와 Y_j가 독립적인 베르누이 p 확률변수이다.  
 이때 X+Y는 모든 j에 대한 X_j + Y_j의 합이다.  
 즉, n+m개의 i.i.p.한 확률변수의 합이다. = > Bin(n+m , p)

 X = j라고 해서 Y의 값을 짐작할 수 없기에 독립이며, 조건부확률의 X=j를 소거한 것임.  
 그리고 P(Y = k-j)를 이항확률변수의 확률질량함수로 표현하였다.
   
  

이제 이항분포가 아닌데 이항분포라고 생각하는 경우의 예제를 보자.  
  
시행이 독립적이고, 성공 확률이 동일해야한다는 것이다.  
그렇지 않다면 이항분포가 아니다.  
  
ex) 일반적인 52장의 카드에서 5장의 카드를 뽑는다고 하자.  
그리고 꺼낸 카드 중 에이스 카드의 개수의 확률분포를 구해보자.  
  
52장의 표본공간 중 크기 5의 모든 부분집합은 동등한 확률로 분포하고,  
에이스의 개수는 어떠한 값을 지닌다.  
이 값은 이산적이기 때문에 확률질량함수를 구하는 것이 좋다.   
X = (# of Aces) 라고 하자.  
  
P(X = k)를 찾아야 한다. k ∈ {0,1,2,3,4} 가 아닌 이상 확률은 0의 값을 지닌다.  
  
그리고 이 분포는 이항분포가 아니다!
