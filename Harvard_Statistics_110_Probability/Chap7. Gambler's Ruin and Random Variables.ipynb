### Gambler's Ruin(도박꾼의 파산) 문제

Gambler A, B가 있다.  
서로 1달러씩 베팅을 하고 같은 게임을 반복해서 하고 있다는 것을 가정한다.  
(각각의 게임은 독립적이며 1달러씩 베팅하는 게임 한 판을 1라운드라 한다.)
둘 중 한명이 파산하면 게임은 끝이 난다.  

p = P(A가 한 라운드에서 이길 확률), q = 1 - p  ( = B가 한 라운드에서 이길 확률)

**여기서 질문은 A가 게임에서 이길 확률(= A가 B를 파산시킬 확률)은?**  
A는 i달러, B는 N - i 달러를 지닌채로 게임을 시작한다.(총 N 달러)  
  
이 문제가 중요한 이유!! :성공 확률과 실패 확률이 정해져 있는 독립적인 사건을 반복하는 문제  
그리고 이 문제를 다르게 생각하면 Random walk(무작위 행보)가 된다.  
-> 문제의 패턴을 알아보자.  
![](https://cphinf.pstatic.net/mooc/20180829_269/1535520674108Cetpy_PNG/7-2.png)  
p : 오른쪽으로 갈 확률,  q : 왼쪽으로 갈 확률  
처럼 어느 방향으로나 갈 수 있는 무작위 행보이다.  
  
그리고 i가 0이나 N에 도달하면 게임이 끝나는 것이다.  
  
i가 왔다갔다하면서 게임이 끝나지 않을 수도 있을 것인데,  
게임이 끝나지 않는 확률이 0이 아니라고 한다.
  
그런데 어떻게 0이 아닌지 알 수 있을까?  
  
이 문제는 **재귀적인** 형태  
한 단계 이후에는 초기 조건만 다른 같은 문제들의 형태로 이루어져 있다.  
  
전략 : 첫 단계에 조건을 세우는 것.  
  
P_i = A가 i달러를 가지고 시작했을 때 전체 게임을 이길 확률  이라고 하자.  

전체 확률의 법칙에 따라,  
**P_i = pㆍP_i+1 + qㆍP_i-1(1 ≤ i ≤ N-1)**  
P_0 = 0(A가 파산한 상태로 시작), P_N = 1(B가 파산한 상태로 시작)이다.(경계 조건 boundary condition)  
  
그리고 위와 같이 **재귀적인**형태를 띄는 방정식을 **계차방정식(difference equation)**이라고 한다.  
계차방정식은 미분방정식의 이산 형태(연속적인 시간에 대해서는 표현 불가)이다.  
  
미분 방정식을 풀 때 처럼 추측(guessing)을 통해 계차방정식을 풀어보자.  

P_i = pㆍP_i+1 + qㆍP_i-1에서  
**P_i = x^i라고 추측해보자.**  
x^i = pㆍx^i+1 + qㆍx^i-1  (x = 0 이 아니라고 가정하고 x를 나눠준다.)  
pㆍx^2 - x + q = 0 
![](https://latex.codecogs.com/gif.latex?x%20%3D%20%5Cfrac%20%7B%20-1%20%5Cpm%20%5Csqrt%7B1%20-%204pq%7D%7D%7B2p%7D)

1-4pq = 1-4p(1-p) = 4p^2 - 4p + 1 = (2p - 1)^2  가 되기 때문에  
![](https://latex.codecogs.com/gif.latex?x%20%3D%20%5Cfrac%20%7B%20-1%20%5Cpm%20%5Csqrt%7B1%20-%204pq%7D%7D%7B2p%7D%20%5Cin%20%5C%7B1%2C%20%5Cfrac%7B2%20-%202p%7D%7B2p%7D%20%3D%20%5Cfrac%7Bq%7D%7Bp%7D%20%5C%7D)  
  
하지만 이것은 경계조건을 만족하지 않는다.
미분방정식을 풀 때 일반해를 구하고 그 다음 초기 조건이나 경계 조건을 사용해서 특수해를 구한다.  
  
우리가 볼 일반해는  
먼저 p ≠ q 일경우 (어떤 한 참가자가 이길 확률이 더 높은 불공평한 게임) 
P_i = Aㆍ1^i + B(q/p)^i  
P_0 = 0 을 대입하면, B = -A가 되며,  
P_N = 1 을 대입하면, 1 = A(1 - q/p)^N 이 된다.  
A = 1/(1 - q/p)^N  
   
그리고 p = q 일경우  (참가자의 이길 확률이 동등한 공평한 게임)  
위의 x = { 1, q/p = 1} = 1 이기 때문에 중복을 없애주기 위해 계차방정식이 더 복잡해짐.( = 1이 중복도 2를 지님)  
x = q/p 라고 하고 x=1일 극한을 봤을 때,  
![](https://latex.codecogs.com/gif.latex?%5Clim_%7Bx%20%5Crightarrow%201%7D%5Clarge%5Cfrac%7B1-x%5Ei%7D%7B1-x%5EN%7D%20%3D%20%5Clim_%7Bx%20%5Crightarrow%201%7D%5Clarge%5Cfrac%7Bi%28x%5E%7Bi-1%7D%29%7D%7BN%28x%5E%7BN-1%7D%29%7D%20%3D%20%5Cfrac%7Bi%7D%7BN%7D)
(by 로피탈의 정리)
 ![](https://latex.codecogs.com/gif.latex?P_%7Bi%7D%20%3D%20%5Cbegin%7Bcases%7D%20%26%20%5Cfrac%20%7B%201%20-%20%7B%5Cfrac%7Bq%7D%7Bp%7D%7D%5E%7Bi%7D%20%7D%7B1%20-%20%7B%5Cfrac%7Bq%7D%7Bp%7D%7D%5E%7BN%7D%7D%5C%20%2C%5C%20%7B%20if%20%7D%5C%20p%5Cneq%20q%20%5C%5C%20%26%20%5Cfrac%20%7Bi%7D%7BN%7D%5C%20%2C%5C%20%7B%20if%20%7D%5C%20p%20%3D%20q%20%5C%5C%20%5Cend%7Bcases%7D)  
  
p=q일 때는 직관적으로 기억하기 쉽다  = 갖고 있는 재산 / 총 재산 이기 때문  
  
p≠q일 경우는  
p = 0.49 (A가 약간 불리할 경우)  
N = 20 -> 0.4  
N = 100 -> 0.12  
N = 200 -> 0.02의 확률로 A가 이긴다.  

**즉, 약간의 불리한 확률을 가지면 시행이 많아질수록 확률이 떨어진다!!**  
  
여기서 게임이 무한정 반복되는 경우는 배제하고 생각한 것일까?  
**사실은 배제한 것이다.**  
  
질문을 B가 이길 확률으로 바꾼다면?
p = q 일경우 A 가 이길확률 i / N 을 통해서 B가 이길 확률은 N - i / N이 될 것이고  
i / N + N - i / N = 1이 되므로  
**게임이 영원히 지속될(즉, A와 B가 아무도 이기지 못하는) 확률은 0이 되는 것이다.**  
  
  
  

### 확률변수(Random variables)  
  
확률변수는 실제로는 함수이다. 함수는 결정되어 있는데 
여기서 random(무작위성)은 어디서 오는 것일까?  
  
확률변수는 S라고 불렀던 표본공간(Samplespace) 안의 함수이다.  
실수(R)에 대한 확률 시행이 있는 것이다.

입력값으로는 결과 중 특정 값인 s을 지니며, 실수값에 대응하게 되는데 보통 확률변수는  X라고 한다.  
  
확률변수는 표본 공간에서 실수로 가는 함수이다.  
![](https://cphinf.pstatic.net/mooc/20180829_53/1535526243836RIiXb_PNG/7-4.png)  
  
여기서는 확률변수를 확률 시행(experiment)의 일부분의 수치적인 '요약(summary)'라고 생각할 것이다. 무작위성은 확률 시행(experiment)에서 오며 이러한 시행은 각 다른 결과들에 대해서 다른 확률을 부여해준다.  
확률 시행이 끝난 뒤 특정한 결과를 받게 되는데, s라고 하는 결과를 시행 후에 관측하는 것이다. 그리고 그것을 실수로 대응시킨다. s는 추상적이지만 실수는 우리가 알기 쉽다.  
  
ex) 베르누이 분포(Bernoulli distribution)  
Defn : 확률변수 X가 베르누이 분포(=Bernoulli_distribution(p))를 따른다고 할 때, X는 0과 1 오직 두가지의 값만 가질 수 있으며,  
P(X=1) = p , P(X=0) = 1- p 이면 X = Bernoulli_distribution(p)가 된다.
즉 확률시행 이후에는 X는 0 또는 1의 값을 가지게 된다.  
   
이항 분포(binomial distribution)     
Binomial(n, p) : n개의 독립적인 Bernoulli_distribution(p)(베르누이 시행) 시도 동안에 x번의 성공횟수의 분포(1 : 성공, 0: 실패)  
  
분포 : 확률변수가 어떤 값을 지닐 지에 대한 청사진

n번 시행하게 되며 x 는 0과 n사이의 정수가 됩니다.  

n = 7이고, k = 3이라고 봤을 때,
p ( x = 3 ) =  7 C 3 ㆍ p^3 ㆍ (1-p)^4

P( x = k ) = n C k ㆍ p^k ㆍ (1-p)^(n-k) , 0 ≤ k ≤ n, k는 정수.( = x=k일 확률 )   
이며 이를 PMF(probability mass function) 확률질량함수 라고 한다.  

이항분포의 특징  
X ~ Binomial(n, p) 이고, Y ~ Binomial(m, p)이고 서로 독립적일 때,  
X+Y ~ Binomial(n+m, p)이다.  

증명 : X라는 처음 n개의 베르누이 시행에 대한 성공과, Y라는 다음 m개의 시행에 대한 성공은 서로 독립적이기에 총 성공횟수를 더하면 X+Y가 되고 정의상 이항 분포가 된다.
